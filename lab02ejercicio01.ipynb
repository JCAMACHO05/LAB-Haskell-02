```haskell
-- EJERCICIO 1
-- 1. Definir las siguientes funciones usando recursión.
-- Retorna la longitud de una lista
-- miLength :: [a] -> Int

-- Longitud de una lista
miLength :: [a] -> Int
miLength []     = 0
miLength (_:xs) = 1 + miLength xs

-- Retorna el último elemento de una lista
-- miLast :: [a] -> a

-- Último elemento de una lista
miLast :: [a] -> a
miLast [x]    = x
miLast (_:xs) = miLast xs

-- Duplica cada elemento de una lista de números
-- duplicar :: [Int] -> [Int]

-- Duplicar cada elemento
duplicar :: [Int] -> [Int]
duplicar []     = []
duplicar (x:xs) = (2*x) : duplicar xs

-- Eliminar los elementos pares de una lista
-- eliminarPares :: [Int] -> [Int]

-- Eliminar elementos pares
eliminarPares :: [Int] -> [Int]
eliminarPares [] = []
eliminarPares (x:xs)
    | odd x     = x : eliminarPares xs
    | otherwise = eliminarPares xs

-- 2. Defina estas mismas funciones pero ahora usando funciones de orden superior como map y filter

-- Longitud usando composición
miLength :: [a] -> Int
miLength = sum . map (const 1)

-- Último elemento (sin recursión explícita)
miLast :: [a] -> a
miLast = foldl1 (\_ x -> x)

-- Duplicar cada elemento
duplicar :: [Int] -> [Int]
duplicar = map (*2)

-- Eliminar elementos pares
eliminarPares :: [Int] -> [Int]
eliminarPares = filter odd
```
